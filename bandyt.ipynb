```python
from transformers import Tacotron2ForSpeech, Tacotron2Processor

processor = Tacotron2Processor.from_pretrained("tacotron2")
model = Tacotron2ForSpeech.from_pretrained("tacotron2")

input_text = "Hello, this is a TTS model."
inputs = processor(text=input_text, return_tensors="pt")

# Generate speech
outputs = model.generate(**inputs)

# Convert to audio
audio = processor.batch_decode(outputs, return_tensors="pt")
```
